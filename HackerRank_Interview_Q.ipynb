# How to find Factorial

```python
num = 7
factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   for i in range(1,num + 1):
       factorial = factorial*i
   print("The factorial of",num,"is",factorial)
```

```python

def factorial(x):
    """This is a recursive function
    to find the factorial of an integer"""

    if x == 1:
        return 1
    else:
        # recursive call to the function
        return (x * factorial(x-1))


# change the value for a different result
num = 7


# call the factorial function
result = factorial(num)
print("The factorial of", num, "is", result)
```

# Program to check if a number is prime or no

```python


num = 2

# To take input from the user
#num = int(input("Enter a number: "))

# define a flag variable
flag = False

if num == 1:
    print(num, "is not a prime number")
elif num > 1:
    # check for factors
    for i in range(2, num):
        if (num % i) == 0:
            # if factor is found, set flag to True
            flag = True
            # break out of loop
            break

    # check if flag is True
    if flag:
        print(num, "is not a prime number")
    else:
        print(num, "is a prime number")
```

# Program to display the Fibonacci sequence up to n-th term

```python


nterms = int(input("How many terms? "))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if nterms <= 0:
   print("Please enter a positive integer")
# if there is only one term, return n1
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
# generate fibonacci sequence
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```

```python
# Python program to display the Fibonacci sequence

def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = 9

# check if the number of terms is valid
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(recur_fibo(i))
```

# without using the sort() function

```python

#Python program to print a list
# without using the sort() function
# without an extra variable
 
l1=[76, 23, 45, 12, 54, 9]
print("Original List:", l1)
 
# sorting list using nested loops
for i in range(0, len(l1)):
    for j in range(i+1, len(l1)):
        if l1[i] >= l1[j]:
            l1[i], l1[j] = l1[j],l1[i]
 
# sorted list
print("Sorted List", l1)
```

```python
# importing libraries
import time
import math

# decorator to calculate duration
# taken by any function.
def calculate_time(func):

    def inner1(*args, **kwargs):
        begin = time.time()
        func(*args, **kwargs)
        end = time.time()
        print("Total time taken in : ", func.__name__, end - begin)
    return inner1



# this can be added to any function present,
# in this case to calculate a factorial
@calculate_time
def factorial(num):
    print(math.factorial(num))

# calling the function.
factorial(10)
```

# list sorting




```python
# 1 Insertion Sort Algorithm
A = [3, 8, 5, 4, 1, 9, -2]
for i in range(1, len(A)):
 
    value = A[i]
    j = i
 
        # find index `j` within the sorted subset `A[0…i-1]`
        # where element `A[i]` belongs
    while j > 0 and A[j - 1] > value:
        A[j] = A[j - 1]
        j = j - 1
 
        # Note that sublist `A[j…i-1]`is shifted to
        # the right by one position, i.e., `A[j+1…i]`
 
    A[j] = value
A
```

```python
A = [3, 8, 5, 4, 1, 9, -2]
for i in range(1, len(A)):
    value = A[i]
    j = i
    
    while j > 0 and A[j-1] > value:
        A[j] = A[j-1]
        j = j-1
    A[j] = value
A
```

# Second highest number without sorting

```python
A = [3, 9, 5, 8, 0]# 4, 1, 900, -2]
fistMax = A[0]
SecondMax = A[0]
for i in range(1,len(A)):
    if A[i] > fistMax:
        SecondMax = fistMax
        fistMax = A[i]
    elif A[i] > SecondMax and fistMax != A[i]:
        SecondMax = A[i]
    elif fistMax == SecondMax and SecondMax != A[i]:
        SecondMax = A[i]
        
        
            
            
print(f'The first hight number is {fistMax} and second highest num is {SecondMax}')   
```
